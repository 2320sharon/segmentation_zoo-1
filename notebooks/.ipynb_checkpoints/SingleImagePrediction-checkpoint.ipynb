```python
#load up the basics
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

#If you have multiple GPU, set which one to use. 
# Numbering starts at 0
os.environ["CUDA_DEVICE_ORDER"] = "PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

#import TF
import tensorflow as tf

#Check TensorFlow Version
print('TF version: {}' .format(tf.__version__))

#Check for GPU utilization
if tf.test.gpu_device_name():
    print(' GPU: {}'.format(tf.test.gpu_device_name()))
else:
    print("No GPU")
```

```python
#Load up the TF model
# here i am using a model trained to predict NOAA post storm images.
# it segments an RGB image into 4 classes (water, sand, vegetation, development)


#model path
model_path = './models/NOAA_NewLoss_fullmodel_model'

# load model into tf
model = tf.keras.models.load_model(model_path, compile = True)
model.summary()
```

```python
#load up an image
ImPath = "../data/images/C26047791-1_EBG.jpg"
img = matplotlib.image.imread(ImPath)
imgplot = plt.imshow(img)
```

```python
#images are standardized before they are sent through the model:
#adjusted by standard deviation, as required by Gym model

def standardize(img):

    N = np.shape(img)[0] * np.shape(img)[1]
    s = np.maximum(np.std(img), 1.0 / np.sqrt(N))
    m = np.mean(img)
    img = (img - m) / s
    del m, s, N
    #
    if np.ndim(img) == 2:
        img = np.dstack((img, img, img))

    return img
```

```python
## And we convert he image to the right format and shape, 
# and then send it through the model

#model has a specific target size.. you can look at the model.summary() above.
im_targetsize = (768,1024)

TestIm = tf.keras.preprocessing.image.load_img(ImPath,target_size = im_targetsize)
TestIm = tf.keras.preprocessing.image.img_to_array(TestIm)
TestIm = standardize(TestIm)
TestIm = np.expand_dims(TestIm,axis=0)
print(TestIm.shape)
TestPred = model.predict(TestIm)
print(TestPred.shape)

# you can see the image is a 1,768,1024, 3 tensor 
#(1 image, 768 Height, 1024 length, 3 channels (RGB))

# and the output is a 1,768,1024, 4 tensor 
#(1 image, 768 Height, 1024 length, 3 channels (1 for each class))
```

```python
#So what are in the channels? these are the output softmax scores
#first lets get rid of the extra dimension
sq_TestPred = TestPred.squeeze()
```

```python
#lets look at the softmax scores for each class.. 

#This is the water score for each pixel
plt.imshow(sq_TestPred[:,:,0])
cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
```

```python
#This is the sand score for each pixel
plt.imshow(sq_TestPred[:,:,1])
cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
```

```python
#This is the vegetation score for each pixel
plt.imshow(sq_TestPred[:,:,2])
cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
```

```python
#This is the development (roads/houses) score for each pixel
plt.imshow(sq_TestPred[:,:,3])
cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
```

```python
# to get the composite label, we need to take the argmax of these channels

#look at the argmaxed labels vs the image
# water (0), sand (1), vegetation (3), dev(4)

lab = np.argmax(sq_TestPred,-1)

plt.imshow(lab)
cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
plt.show()
plt.imshow(img)
```

```python
# you can see that this basic prediction is definitely not perfect. 
#Can we find specific areas in the picture where the softmax scores for each pixel are very close,
# i.e., areas where the model is most confused?
```

```python
# now the next goal is to find the top 2 predictions at each pixel
# we use sort for that
sortedPreds = np.sort(sq_TestPred)
TopPred = sortedPreds[:,:,3]
SecondPred = sortedPreds[:,:,2]

#make an array of confidence. A perfect sacore is 1, so the difference between 1 and the 
#top score is a measure of confidence. More confident areas are closer to 0

LConfidence = 1 - TopPred


#the best pred will have a smallest number. print the mean over the image
print(LConfidence.mean())

#and show trhe image
plt.imshow(LConfidence)

cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
plt.show()
```

```python
#Fropm the above prediction, the model is very confident in itself, except at the edges
```

```python
#make an array of the margin. A perfect score is 1, so the difference between the top class and the next class
# should be 1 (at best). Areas closer to 0 are places where model is confused

Margin = (TopPred - SecondPred)

#the worst pred will have a smallest Margin
#print the mean over the image
print(Margin.mean())

#and show an image
plt.imshow(Margin)

cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
plt.show()
```

```python
#From the above prediction, the model has low margins between top 2 predictions, except at the edges
```

```python
#One way to try and make predictions better is to use test Time augmentation, 
# where we show the model flipped version of the same image, 
#the then average the precdtions to amke a final predcitons

#Original precition
TestIm = tf.keras.preprocessing.image.load_img(ImPath,target_size = im_targetsize)
TestIm = tf.keras.preprocessing.image.img_to_array(TestIm)
TestIm = standardize(TestIm)


#pred1
normal = np.expand_dims(TestIm,axis=0)
pred1 = model.predict(normal).squeeze()

#pred2

flip = np.expand_dims(np.flipud(TestIm),axis=0)
pred2 = np.flipud(model.predict(flip).squeeze())

#flip 2 
flip = np.expand_dims(np.fliplr(TestIm),axis=0)
pred3 = np.fliplr(model.predict(flip).squeeze())


#flip 3

flip = np.expand_dims(np.flipud(np.fliplr(TestIm)),axis=0)
pred4 = np.flipud(np.fliplr(model.predict(flip).squeeze()))


est_label = (pred1 + pred2 + pred3 + pred4) / 4

TTAlab = np.argmax(est_label,-1)

#plot the TTA label
plt.imshow(TTAlab)
cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
plt.show()

#the orioginal label froma  single prediction
lab = np.argmax(pred1,-1)

plt.imshow(lab)
cax = plt.axes([1, 0.1, 0.075, 0.8])
plt.colorbar(cax=cax)
plt.show()

# and the orgiinal image
plt.imshow(img)


```

```python
#The first label (from TTA), is better than the 2nd label (without TTA)
```
